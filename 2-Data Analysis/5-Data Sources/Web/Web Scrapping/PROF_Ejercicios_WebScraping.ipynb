## Ejercicio 2: Encuentra todas las etiquetas de enlace y enumera las diez primeras de la p√°gina web python.org

```python
print("First four h2 tags from the webpage python.org.:")
print(soup.find_all('a')[0:10])
print("\n")
for x in soup.find_all('a')[0:10]:
    print(x["href"], end="\n"*2)
```

```python
# Imponemos que sean http o https
print("First four h2 tags from the webpage python.org.:")
print(soup.find_all('a', attrs={"href": lambda x: "http" in x}, limit=10))
print("\n")
for x in soup.find_all('a', attrs={"href": lambda x: "http" in x}, limit=10):
    print(x["href"], end="\n"*2)
```

```python
i=0
lim=10
for x in soup.find_all('a'):
    if 'http' in x['href']:
        print(x["href"], end="\n"*2)
        i+=1
    
    if i == lim:
        break    
```

```python
def get_forex(exchange, month, year):
    '''
    Funci√≥n que construye la consulta sobre las tablas forex.

    Input:
        - exchange: ['EURUSD', 'GBPUSD', 'AUDUSD', 'NZDUSD', 'USDCAD', 'USDCHF', 'USDJPY', 'USDSEK']
        - month: int (d)
        - year: int (yyyy)
    Output:
        - Pandas DataFrame conn la informaci√≥n
    '''
    
    url_base = 'https://www.myfxdata.io/priceaction/daily.php'
    options = f'D1_Pares={exchange}&D1_Anyos={year}&D1_Meses={month}'

    url = '?'.join([url_base, options])
    # print(url)
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'lxml')

    ls_rows = soup.find('table', class_="sample").contents
    clean_rows = [x for x in ls_rows if type(x) == bs4.element.Tag]

    ls_df = []
    for row in clean_rows:

        cl_row = [x.text for x in row.contents if x!="\n"]
        ls_df.append(cl_row)

    df = pd.DataFrame(data=ls_df[1:], columns=ls_df[0]).dropna()
    return df
```

```python
from itertools import product
import pandas as pd

months = range(1, 13)
years = range(2010, 2022)
exchange = ['EURUSD', 'GBPUSD', 'AUDUSD', 'NZDUSD', 'USDCAD', 'USDCHF', 'USDJPY', 'USDSEK']

ls_args = product(exchange, months, years) #[list(x) for x in product(exchange, months, years)]

df_ls = []
for exc, month, year in ls_args:
    # print(args)
    try:
        df_for_i = get_forex(exc, month, year)
        df_for_i["Exchange"] = exc
        df_ls.append(df_for_i)
        # print(df_for_i)
    except NameError:
        pass

df_forex = pd.concat(df_ls, ignore_index=True)
df_forex
```
